**일괄 처리 함수 만들기**

```python
def data_cleaning(filename):

  ns_df = pd.read_csv(filename, low_memory=False) # 파일을 데이터프레임으로 읽기
  ns_book = ns_df.dropna(axis=1, how='all') # NaN 열을 삭제
  count_df = ns_book[['도서명','저자','ISBN','권','대출건수']] # 대출건수 합치기 위해 필요한 행만 추출하여 count_df 데이터프레임을 만듦
  loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'], dropna=False).sum() # 도서명, 저자, ISBN, 권을 기준으로 대출건수를 groupby 함

  dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권']) # 원본 데이터프레임에서 중복된 행을 제외하고 고유한 행만 추출하여 복사
  unique_rows = ~dup_rows
  ns_book3 = ns_book[unique_rows].copy()

  ns_book3.set_index(['도서명','저자','ISBN','권'], inplace=True) #도서명, 저자, ISBN, 권을 인덱스로 설정
  ns_book3.update(loan_count) # loan_count에 저장된 누적 대출건수를 업데이트함
  ns_book4 = ns_book3.reset_index() # 인덱스 재설정
  ns_book4 = ns_book[ns_book.columns] # 원본 데이터프레임의 열 순서로 변경
  return ns_book4
```

```python
new_ns_book4 = data_cleaning('ns_202104.csv')
# ns_book4.equals(new_ns_book4) # 다른 데이터프레임을 비교할때 eqals() 메서드 사용
print(ns_book4.head())
print(new_ns_book4.head())

#여기서 대출건수가 float형 int형으로 다르게 저장되었는데 그 이유를 모르겠음...
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def get_book_title(isbn):
  url='https://www.yes24.com/product/search?query={}'
  r=requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser')

  title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  return title
```

저자, 출판사, 발행 연도를 추출하여 반홯하는 함수</br>

저자는 두 명 잇상일 수 있기 때문에 find_all() 메서드 사용</br>
'2020년 12월'처럼 쓰여 있으면 정수만 추출해야하기 때문에 re 모듈의 findall() 함수를 사용

```python
import re

def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub = row['출판사']
  year = row['발행년도']

  url ='https://www.yes24.com/product/search?query={}'
  r = requests.get(url.format(row['ISBN']))
  soup = BeautifulSoup(r.text, 'html.parser')

  try:
    if pd.isna(title):
      title=soup.find('a', attrs={'class':'gd_name'}).get_text()
  except AttributeError:
    pass

  try:
    if pd.isna(author):
      authors = soup.find('span', attrs={'class':'authPub info_auth'}).find_all('a')
      authors_list= [auth.get_text() for auth in authors]
      author = ','.join(authors_list)
  except AttributeError:
    pass


  try:
    if pd.isna(pub):
      pub = soup.find('span', attrs={'class':'authPub info_pub'}).find_all('a').get_text()
  except AttributeError:
    pass


  try:
    if year==-1:
      year_str = soup.find('span', attrs={'class':'authPub info_data'}).get_text()
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass

  return title, author, pub, year
```
